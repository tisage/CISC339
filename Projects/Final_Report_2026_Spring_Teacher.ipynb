## Part I — GTC 2026: The Industry View

**Q1.** Jensen describes AI as having gone through **three major inflection points**: (1) ChatGPT in 2022, (2) reasoning models like o1 in 2024, and (3) agentic AI in 2025. Describe each transition in your own words. Why does Jensen argue that the third shift — agentic AI — is the most consequential for computing demand?

**Model Answer:**

The three inflection points Jensen describes are:

1. **ChatGPT (2022)** — AI could perceive and generate language for the first time at a level that felt genuinely useful. This launched the era of generative AI and showed the world that conversational AI was viable.

2. **Reasoning models / o1 (2024)** — AI became "credible, trustworthy, and reliable" by thinking through problems step by step before responding. Rather than just generating fluent text, models could now reason, check their work, and produce more reliable outputs.

3. **Agentic AI / Claude Code (2025)** — AI can now be given a high-level goal and autonomously plan, use tools, read files, write code, test results, and iterate. Jensen's own words: *"You don't ask an AI what, where, when, how. You ask it create, do, build."*

Agentic AI is most consequential for compute because agents work continuously and generate massive amounts of inference-time compute — not one response to one question, but thousands of reasoning steps per task, running 24/7. This is why Jensen claims computing demand has increased by "1 million times in the last two years" and why he projects at least $1 trillion in demand through 2027.

**Key quote (GTC transcript, ~48:40):** *"An AI that was able to perceive became an AI that could generate. An AI that could generate became an AI that could reason. An AI that could reason now became an AI that can actually do work."*

*Grading note: Award full credit if student correctly identifies all three stages and explains that agents create ongoing inference demand rather than one-time responses.*

**Q2.** NVIDIA announced **Vera Rubin** as the successor to Blackwell, explicitly describing it as "architected for every phase of Agentic AI." What technical capabilities does Jensen highlight, and what does the phrase "7 chips, 5 rack-scale computers" suggest about how NVIDIA thinks about computing systems today? Why does Jensen argue that demand for faster hardware will continue to grow rather than plateau?

**Model Answer:**

Vera Rubin delivers **3.6 exaFLOPS** of compute and **260 TB/s** all-to-all NVLink bandwidth, packaged across 7 chips and 5 rack-scale computers. The "7 chips, 5 rack-scale computers" framing signals that NVIDIA no longer thinks of a product as a single chip — the entire rack (CPU, GPU, LPU, networking, storage, security) is now designed and sold as one integrated system. Jensen calls this "vertically integrated, horizontally open."

Jensen argues demand will not plateau because three simultaneous scaling laws are all running at once: **post-training** (RLHF, fine-tuning gets more expensive as models improve), **inference-time compute** (reasoning models think longer for harder problems), and **agentic compute** (agents run continuously on long multi-step tasks). As token costs fall, businesses deploy more AI, creating more demand. Jensen notes that in 10 years NVIDIA has achieved **40 million times more compute** and that each generation funds the next through customer ROI.

*Grading note: Students do not need to know the exact specs. Credit for understanding that Vera Rubin is a system-level product designed for agentic workloads, and for identifying the demand-side argument (at least two of the three scaling laws).*

The uncertainty has shifted from "is this possible?" to "how fast do we scale?"

He also announced a deployment partnership with **Uber** across multiple cities. NVIDIA's strategy is not just to sell GPUs for training the AI models — it is to embed its full computing platform *inside* the vehicles (the NVIDIA DRIVE system) and *inside* the data centers running continuous re-training and fleet management. NVIDIA becomes essential infrastructure at every layer of the autonomous vehicle stack.

**Q4.** Jensen describes the evolution of AI as: *"An AI that could perceive became an AI that could generate. An AI that could generate became an AI that could reason. An AI that could reason now became an AI that can actually do work."* He introduces **OpenClaw** as the "operating system of agent computers," comparing it to what Windows was for PCs. What is an AI agent, and how does it differ from the chatbot-style AI you used in earlier labs? Give one concrete example from the keynote.

**Model Answer:**

A **chatbot** receives a question and returns a single response. An **AI agent** receives a high-level goal and autonomously plans, uses tools, reads files, writes and runs code, reflects on the results, and iterates — all without a human directing each step. Jensen summarizes the shift: you no longer ask AI *"what, where, when, how"* — you ask it to *"create, do, build."*

The concrete example from the keynote is **OpenClaw** (and its enterprise counterpart NemoClaw): a software agent framework where you give the AI a programming task and it breaks it into subtasks, writes code, executes it, debugs failures, and delivers a working result autonomously. Jensen compares OpenClaw to what Windows was for personal computers — it is the platform that makes it possible to build and deploy individual AI agents, just as Windows made it possible to build and run individual applications. He also described NemoClaw adding a security layer ("Open Shell") for enterprise environments.

*Grading note: Core distinction — chatbot responds to prompts, agent executes multi-step tasks autonomously using tools. Credit any concrete example from the keynote (not just OpenClaw — robotaxis, Cosmos robots, etc. also qualify).*

**Q5.** Jensen projected **"at least $1 trillion"** in AI computing demand through 2027 — up from his $500 billion prediction at GTC 2025 just one year earlier. He also claimed: *"I believe that computing demand has increased by 1 million times in the last two years."* What reasoning does he offer to justify these figures? He points to three simultaneous "scaling laws" driving compute demand — what are they?

**Model Answer:**

Jensen's reasoning: last year he saw "very high confidence demand and purchase orders" of $500 billion through 2026. A year later, with AI adoption accelerating across enterprises and governments, he now sees at least **$1 trillion through 2027** — and explicitly said he believes the true number will be higher. His framing is not speculative — he ties it to actual committed orders visible in NVIDIA's revenue pipeline.

The three simultaneous scaling laws he identifies are:
1. **Post-training scaling** — the compute needed to fine-tune, RLHF, and align models keeps growing as models improve.
2. **Inference-time compute scaling** — reasoning models (like o1) think longer for harder problems, consuming more compute per query.
3. **Agentic compute scaling** — AI agents run continuously on multi-step tasks, generating far more inference demand than single-turn chatbots.

All three compound simultaneously, which is why Jensen argues "1 million times more demand in two years" is not hyperbole — each law individually would be significant; all three together create exponential growth.

*Grading note: The specific $1T justification and the three scaling laws are the key content points. Award partial credit if student identifies at least two of the three laws.*

## Part II — Stanford AI Index 2026: The Data View

**Q7.** The report introduces the concept of the **"jagged frontier"** of AI capability. Define this term in your own words and illustrate it with one specific example from the report. Why does this concept matter when evaluating confident claims about what AI can or cannot do?

**Model Answer:**

The "jagged frontier" describes how AI capability is **highly uneven** — AI can be superhuman on some tasks and surprisingly poor on others, with no simple rule predicting which. It is not a smooth curve where harder tasks are uniformly harder; the performance landscape is jagged and counterintuitive.

The Stanford report's clearest example: *"AI models can win a gold medal at the International Mathematical Olympiad but cannot reliably tell time."* Gemini Deep Think solved IMO-level math problems (performance far exceeding most humans) yet leading models cannot accurately read an analog clock face — a task a 6-year-old can perform.

This concept matters because it warns against sweeping claims like "AI is now smarter than humans" or "AI can do X, therefore it can do Y." When Jensen Huang says "the ChatGPT moment of self-driving cars has arrived," the jagged frontier suggests caution: a system that navigates highways reliably may fail unpredictably in construction zones, heavy rain, or unusual pedestrian behavior. Capability on benchmark tasks does not guarantee capability in all real-world conditions.

*Grading note: Requires both a definition and a specific example. Credit any valid example from the report — the time-reading/IMO contrast is the most vivid but not the only one.*

**Q8.** Stanford documents a striking divide: **73% of AI experts** expect AI to have a positive impact on how people do their jobs, compared to only **23% of the general public** — a 50-point gap. What factors might explain this divide? Which group do you think has a more accurate picture of reality, and why?

**Model Answer:**

Several factors explain the gap. Experts interact with AI systems daily in controlled, high-capability contexts and tend to focus on AI's best performance. They also have more economic security and are less likely to hold jobs at high displacement risk. The general public experiences AI primarily through media coverage that emphasizes job loss fears, high-profile failures, and deepfakes — rather than productivity gains. People naturally worry more about losing their current situation than they optimistically project future improvements.

There is a reasonable argument for each side: experts may be subject to insider optimism bias (similar to how tech industry workers consistently overestimate adoption of their own products). The public may be subject to loss aversion and media-driven fear. The Stanford data itself provides a nuance — entry-level software developer employment has fallen 20%, which suggests the public's concern is not baseless, even if the expert assessment of long-term positive impact may also be correct.

*Grading note: This is an open-ended question. Full credit for clearly explaining the gap (at least 2 factors) and taking a supported position. Penalize answers that simply say "experts are right" or "the public is right" without reasoning.*

**Q9.** The report documents **productivity gains of 14–26%** in fields like software development and customer support — yet employment among young software developers (ages 22–25) fell nearly **20%**. How do you reconcile these two findings? What does this suggest about the relationship between AI-driven productivity and job availability?

**Model Answer:**

These findings are not contradictory — they represent a well-documented historical pattern called the **productivity-employment paradox**. When tools make each worker more productive, companies can produce the same output with fewer workers. The gains accrue primarily to firms (lower costs, higher margins) and to senior workers who use AI to amplify their output, while entry-level roles that are most easily automated are reduced or eliminated.

In software development specifically, AI coding assistants (GitHub Copilot, Cursor, etc.) allow one experienced developer to do work that previously required two or three junior developers. The productivity gain is real — but the demand for human labor decreases at the margin. Young workers (22–25) are hit hardest because they typically enter through precisely these junior roles.

This suggests that AI-driven productivity does not automatically create new jobs fast enough to replace those it displaces — at least not in the short term and not for the same workers. The economic gains from productivity require active policy choices (retraining programs, education reform, labor policy) to be broadly shared.

*Grading note: Core insight is the productivity-employment paradox — productivity gains and job losses can coexist. Credit any historically-grounded explanation (automation has always done this) plus any acknowledgment that the effects are not uniform across experience levels.*

**Q10.** Despite leading the world in AI private investment ($285.9 billion in 2025), the United States ranks **24th globally in generative AI adoption** at only 28.3%, while Singapore (61%) and the UAE (54%) rank far higher. What might explain this gap between investment leadership and consumer adoption?

**Model Answer:**

Several factors can explain this paradox:

1. **Investment ≠ adoption**: Most U.S. AI investment goes into model training and infrastructure (data centers, chips) — not consumer-facing deployment. Building GPT-4 does not mean Americans use it.
2. **Regulatory and liability caution**: U.S. enterprises face legal uncertainty around AI use in healthcare, finance, and employment — slowing deployment in regulated sectors.
3. **Market structure**: Singapore and UAE are smaller, more centralized economies where government-led AI adoption campaigns reach the population quickly. The U.S. is a decentralized market of 330 million people across highly varied industries.
4. **Digital infrastructure baseline**: Countries like Singapore already have very high smartphone penetration, digital payment adoption, and tech literacy — which correlates with faster AI tool adoption.
5. **GDP correlation**: Stanford notes adoption correlates strongly with GDP per capita, but the U.S. underperforms even relative to its wealth — suggesting cultural, regulatory, or awareness factors are also at play.

*Grading note: Accept any 2–3 well-reasoned factors. The key insight is that investment and adoption are different phenomena. Penalize superficial answers that only say "the U.S. is big."*

**Q11.** The number of AI researchers and developers moving to the United States has dropped **89% since 2017**, with 80% of that decline occurring in the last year alone. The report frames this as a risk to U.S. AI leadership. Do you agree? What are the broader implications — for the U.S., for other countries, and for the global AI ecosystem?

**Model Answer:**

Yes, this is a significant risk. Historically, U.S. AI leadership has been built partly on its ability to attract the world's best researchers — DeepMind researchers moved to Google, researchers from Chinese universities contributed to U.S. labs, and international PhD students stayed after graduating. An 89% drop in this pipeline means the U.S. is increasingly reliant on domestically trained talent, which is a slower-growing pool.

For the U.S.: the immediate risk is a talent gap at frontier labs and reduced diversity of perspectives in AI research. Longer term, it affects the U.S.'s ability to maintain its lead in novel model development.

For other countries: brain drain reversals (researchers returning to China, India, Europe, South Korea) strengthen their domestic AI ecosystems. This could accelerate AI development outside the U.S., making the global landscape more multipolar — which may actually be positive for global AI governance and diversity of approaches.

For the global AI ecosystem: a more geographically distributed research base may reduce concentration risk and lead to more linguistically and culturally diverse AI systems. However, it also complicates international coordination on safety standards.

*Grading note: The question asks about three levels (U.S., other countries, global ecosystem). Credit answers that address at least two. Look for nuanced thinking — not all consequences of this trend are negative globally.*

**Q12.** Industry now produces **over 90% of notable AI models**, and frontier labs are increasingly withholding training details — parameters, dataset sizes, training code. Why does scientific transparency matter in AI development? Who benefits from less transparency, and who is harmed?

**Model Answer:**

Scientific transparency matters for several reasons: (1) **Reproducibility** — other researchers cannot verify, challenge, or build on results they cannot inspect; (2) **Safety auditing** — regulators, civil society, and independent researchers cannot assess risks in systems they cannot examine; (3) **Accountability** — when a model behaves harmfully, the lack of training data disclosure makes it nearly impossible to understand why; (4) **Power concentration** — opacity entrenches the advantage of incumbents who already have large models, making it harder for new entrants or academic researchers to compete.

**Who benefits from less transparency**: Frontier labs (OpenAI, Anthropic, Google DeepMind) — they protect trade secrets, prevent competitors from replicating their work cheaply, and avoid scrutiny of potentially problematic training data.

**Who is harmed**: Academic researchers (cannot verify or build on the work), policymakers (cannot regulate what they cannot understand), the public (cannot audit systems making consequential decisions about their lives), and safety researchers (cannot find failure modes in closed systems).

*Grading note: Award full credit for addressing both the "why transparency matters" and the asymmetric beneficiary question. Superficial answers ("transparency is good") without specific reasoning should receive partial credit.*

**Q13.** GTC 2026 presents AI as an unstoppable technological wave. The Stanford AI Index presents a more nuanced picture: rapid capability growth alongside uneven adoption, emerging labor disruption, and governance gaps. Which narrative do you find **more useful** for understanding where AI actually stands today? Support your answer with specific evidence from both sources.

**Model Answer (example — strong response):**

Both narratives capture something real, but the Stanford framing is more *useful* for understanding AI's actual current state — not because it is more optimistic or pessimistic, but because it provides falsifiable data rather than projections.

GTC is accurate that AI capability is advancing rapidly: the SWE-bench coding benchmark rose from 60% to near 100% of human baseline in a single year (Stanford, Takeaway 1), consistent with Jensen's claim that computing demand has "increased by 1 million times." These are real developments.

However, Jensen's framing systematically excludes the deployment gap. Stanford shows that despite the capability gains, AI agent deployment "remains in single digits across nearly all business functions" (Takeaway 9). Jensen announces $1 trillion in demand — Stanford shows the U.S. ranks 24th in actual consumer adoption. These are not contradictory — investment and adoption can diverge — but Jensen's narrative treats them as the same thing.

The Stanford report's most useful contribution is forcing the question: *for whom is this progress, and at what pace?* Jensen's keynote is essential for understanding the technology's direction; Stanford is essential for understanding where society actually is.

*Grading note: Full credit requires citing specific evidence from BOTH sources and making a clear, reasoned choice. Neither answer (GTC or Stanford) is inherently correct — evaluate quality of argument.*

**Q14.** Jensen declared *"The ChatGPT moment of self-driving cars has arrived."* Using the Stanford report's concept of the jagged frontier and the documented gap between AI capability and real-world deployment, critically evaluate this claim. Is it a well-supported statement, strategic marketing, or something in between? What evidence would you need to see to fully believe it?

**Model Answer (example — strong response):**

The claim is somewhere between a well-supported milestone and strategic marketing — and the distinction matters.

In favor of the claim: Waymo currently operates commercial robotaxi services in multiple U.S. cities. Major automakers are committing to NVIDIA's platform, suggesting genuine commercial confidence. The analogy is apt in one sense: in 2022, ChatGPT showed that conversational AI *could* work — deployment came later. Jensen may be saying we've crossed that "existence proof" threshold for self-driving.

Against the claim: The Stanford jagged frontier concept warns that demonstrated performance in tested conditions does not guarantee performance across all conditions. ChatGPT works in all languages for all users — a robotaxi must work at 3am in heavy snow near an unmarked construction zone. These are different reliability requirements. Stanford also documents that AI deployment "remains in single digits" across business functions despite capability gains — the gap between "can work" and "widely deployed safely" has historically been large and slow to close.

Evidence needed: Safety statistics across diverse conditions (not just favorable routes), independent (non-NVIDIA) validation of the deployment timeline, and regulatory approval in multiple jurisdictions. A true "ChatGPT moment" would be visible in public adoption data — not just partnership announcements.

*Grading note: Full credit for using both the jagged frontier concept and the deployment gap, and for identifying what specific evidence would change the evaluation.*
